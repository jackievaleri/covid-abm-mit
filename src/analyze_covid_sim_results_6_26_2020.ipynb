```python
# imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
ptrans_tests = [0.01, 0.05, 0.1, 0.2, 0.25] # ptrans = Transmission probability
progression_period_tests = [1, 2, 3, 4, 5, 7] # progression_period = Average number of days until a patient seeks treatment.
interactions_tests = [0.25, 0.5, 1, 2] # interactions = Average number of interactions per person per day (decreases with social distancing)
```

```python

param_total_infections = []

for ptrans in ptrans_tests:
    for progression_period in progression_period_tests:
        for interactions in interactions_tests:
            param_path = str(ptrans) + '_' + str(progression_period) + '_' + str(interactions)
            filename = '../out/' + param_path + '/COVID_output_' + param_path + '.csv'
            curr_df = pd.read_csv(filename)
            total_infected = int(np.max(list(curr_df.loc[:,'recovered'])) + np.max(list(curr_df.loc[:,'dead'])))
            
            param_total_infections.append([ptrans, progression_period, interactions, total_infected])
            
            
df = pd.DataFrame(param_total_infections,
                  columns = ['transmission probability', 
                             'progression period',
                             'interactions per person per day',
                             'total students infected']).pivot_table(values = ['total students infected'],
                                                             columns = ['interactions per person per day'],
                                                             index = ['transmission probability', 'progression period'])
df = df.round(3)
(df.style.background_gradient(cmap=sns.cubehelix_palette(start=.5, rot=-.75, as_cmap=True)))
```
